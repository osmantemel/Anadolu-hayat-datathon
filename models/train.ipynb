```python
eksik_veri_sayisi = data_Train.isnull().sum()

plt.figure(figsize=(20,6))
eksik_veri_sayisi.plot(kind='bar', color='skyblue')
plt.title('Her Kolondaki Eksik Veri Sayısı')
plt.xlabel('Kolonlar')
plt.ylabel('Eksik Veri Sayısı')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
